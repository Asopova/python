```python
#Задание 1**

import pandas as pd
rating = pd.read_csv("ratings.csv")
def claster(data):
    if data['rating'] >= 4.5:
        return("high")
    elif data['rating'] > 2:
        return("average")
    else: return("low")
rating["class"] = rating.apply(claster, axis = 1)
rating.head()
```

```python
#Задание 2

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }
geo = pd.read_csv("keywords.csv")
geo.head()
```

```python
list = [] 
for i in geo_data.values():
    for i in i:
        list.append(i)
def check(keywords):
    for i in keywords:
        if i in list:
            for items in geo_data.items():
                if i in items[1]:
                    print(items[0])
check(geo)
```

```python
def region_in_keyword(row):
    data=row['keyword'].split(' ')
    i=0
    for word in data:
        if word in list:
            for items in geo_data.items():
                if word in items[1]:
                    i+=1
                    return items[0]
    if i==0:
        return 'undefined'
geo["region"] = geo.apply(region_in_keyword, axis=1)
geo[geo["region"]!='undefined']
```

```python
#Задание 3
rating = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
years = [str(i+1950) for i in range(72)] #Столбец с годом расширила до 2022
data = rating.merge(movies, how='left')
data.head()

films_rating = data.groupby('title').mean().reset_index()[['title', 'rating']]

films_rating.head()
def production_year(row):
    for year in years:
        if year in row['title']:
            return year
    return '1900'
films_rating["year"] = films_rating.apply(production_year, axis = 1)
films_rating.head()
films_rating.sort_values('rating', ascending=False).head(50)
```
